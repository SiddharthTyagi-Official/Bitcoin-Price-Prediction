```
g = sns.jointplot( "btc_difficulty", "btc_market_price", data=Bitcoin, xlim=(0, 2.6E+12), ylim=(0, 20000), color="r", size=10)
k = sns.stripplot(x="btc_n_orphaned_blocks",y="btc_market_price",data=Bitcoin,jitter=True)
a = sns.jointplot( "btc_hash_rate", "btc_market_price", data=Bitcoin, xlim=(0, 30000000), ylim=(0, 20000), color="r", size=10)
b = sns.jointplot("btc_market_price","btc_n_transactions", data=Bitcoin, ylim=(0, 400000), xlim=(0, 20000), color="r", size=15)
c = sns.jointplot("btc_market_cap", "btc_market_price", data=Bitcoin, xlim=(0, 450000000000), ylim=(0, 20000), color="r", size=15)
d = sns.jointplot("btc_trade_volume", "btc_market_price", data=Bitcoin, xlim=(0, 1000000000), ylim=(0, 20000), color="r", size=15)
```

```
Bitcoin['btc_trade_volume'].fillna(Bitcoin['btc_trade_volume'].median(),inplace = True)
Bitcoin['btc_transaction_fees'].fillna(Bitcoin['btc_transaction_fees'].median(),inplace = True)
Bitcoin['btc_difficulty'].fillna(Bitcoin['btc_difficulty'].mean(),inplace = True)
```

```
import statsmodels.formula.api as smf
import sklearn.metrics as sm
```

```
model = smf.ols(formula='btc_market_price ~ btc_market_cap + btc_n_transactions + btc_miners_revenue + btc_cost_per_transaction + btc_difficulty + btc_hash_rate + btc_cost_per_transaction_percent', data=Bitcoin).fit()
```

```
y = model.predict(Bitcoin[['btc_market_cap','btc_cost_per_transaction_percent','btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate']])

print("MSE:", sm.mean_squared_error(y, Bitcoin['btc_market_price'].values))
```
